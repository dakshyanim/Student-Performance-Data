Bivariate Analysis

```python
sns.countplot(x="studytime", hue= "G3", data=df)
plt.show()
```

```python
sns.countplot(x="traveltime", hue= "G3", data=df)
plt.show()
```

```python
sns.countplot(x="freetime", hue= "G3", data=df)
plt.show()
```

```python
sns.boxplot(x="failures", y="G3", data=df)
plt.show()
```

```python
sns.boxplot(x="school", y="G3", data=df)
plt.title("Maths Final Grade (G3) by School")
plt.show()
```

```python
sns.boxplot(x="sex", y="G3", data=df)
plt.title("Math Final Grade (G3) by School")
plt.show()
```

```python
Boys performed slightly better than the girls , though results oveerlap
```
